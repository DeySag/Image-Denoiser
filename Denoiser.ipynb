```python
import matplotlib.pyplot as plt
import cv2
```

```python
def denoise_image(image_path):
  noisy_image = cv2.imread(image_path)
  noisy_image_rgb = cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB)

  gaussian_denoised_image = cv2.GaussianBlur(noisy_image,(11,11),0)
  gaussian_denoised_image_rgb = cv2.cvtColor(gaussian_denoised_image,cv2.COLOR_BGR2RGB)

  bilateral_denoised_img = cv2.bilateralFilter(gaussian_denoised_image,35,25,25)
  bilateral_denoised_img_rgb = cv2.cvtColor(bilateral_denoised_img,cv2.COLOR_BGR2RGB)

  plt.figure(figsize=(15,10))

  plt.subplot(1,2,1)
  plt.imshow(noisy_image_rgb)
  plt.title('Noisy Image')
  plt.axis('off')
  plt.subplot(1,2,2)
  plt.imshow(bilateral_denoised_img_rgb)
  plt.title('Gaussian & Bilateral Denoised Image')
  plt.axis('off')
  plt.show()
```

```python
def denoise_snpimg(image_path):
  noisy_image = cv2.imread(image_path)
  noisy_image_rgb = cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB)
  median_denoised_img = cv2.medianBlur(noisy_image,3)
  median_denoised_img_rgb = cv2.cvtColor(median_denoised_img,cv2.COLOR_BGR2RGB)
  plt.figure(figsize=(15,10))
  plt.subplot(1,2,1)
  plt.imshow(noisy_image_rgb)
  plt.title('Noisy Image')
  plt.axis('off')
  plt.subplot(1,2,2)
  plt.imshow(median_denoised_img_rgb)
  plt.title('Median Denoised Image')
  plt.axis('off')
  plt.show()
```

```python
denoise_image('noisy1.jpg')
```
